Question 1
plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)


```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
x=np.arange(40,50)
y=np.arange(50,60)
plt.plot(x,y,'bo--',linewidth=4,markersize=8)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Question 2)
Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2


```python
x=[1,2,3,4,5,6,7] #days of week
y=[160,150,140,145,175,165,180] #sales
plt.plot(x,y,color='y', marker='o')
plt.show()

days = [1,2,3,4,5,6,7] #days of the week 
sales_1 = [160, 150, 140, 145, 175, 165, 180] #sales of the company 1
sales_2 = [70, 90, 160, 150, 140, 145, 175] #sales of the company 2

plt.plot(days, sales_1,label='sales_1',marker='o',color='b')
plt.plot(days, sales_2,label='sales_2',marker='o',color='r')
plt.title('Sales distribution',color='orange')
plt.xlabel('days of the week',color='blue')
plt.ylabel('sales',color='blue')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
